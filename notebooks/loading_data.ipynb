```
# Loading a .csv  with several doble marks "" 
path_file = join(FILES_DIR, "csv_with_several_doble_marks.csv")

n_columns = 9
rm_quote = lambda x: x.replace('"', '')
converter_dic = { '\"{}\"'.format(i) : rm_quote for i in range(n_columns)}

df = pd.read_csv(path_file, 
                 header= None,
                 engine = 'python',
                 #converters=converter_dic ,
                 encoding='latin-1')
df.head()
```

Converters

```
rm_quote = lambda x: x.replace('"', '')

df = pd.read_csv('test.csv', delimiter='; ', engine='python', 
     converters={'\"j\"': rm_quote, 
                 '\"x\"': rm_quote})

df = df.rename(columns=rm_quote)
```
